```python
import os
import glob
import librosa
import kagglehub

# Descargar la última versión del dataset
path = kagglehub.dataset_download("saurabhshahane/mexican-emotional-speech-database-mesd")
print("Path to dataset files:", path)

# Definir la ruta base donde se encuentran los archivos
base_path = path

# Buscar todos los archivos .wav en el directorio
all_files = glob.glob(os.path.join(base_path, "**/*.wav"), recursive=True)

# Emoción a filtrar
target_emotion = "Happiness"

# Filtrar por emoción
emotion_files = [f for f in all_files if target_emotion in os.path.basename(f)]

print(f"Procesando {len(emotion_files)} archivos de audio con emoción '{target_emotion}'...")

# Lista para almacenar resultados
filtered_files = []

# Duración mínima y máxima en segundos
min_duration = 1.0
max_duration = 15.0

# Medir la duración de cada archivo y filtrar
for audio_file in emotion_files:
    try:
        # Cargar el archivo de audio y obtener su duración
        y, sr = librosa.load(audio_file, sr=None)
        duration = librosa.get_duration(y=y, sr=sr)

        # Filtrar por duración
        if min_duration <= duration <= max_duration:
            filtered_files.append((audio_file, duration))
    except Exception as e:
        print(f"Error al procesar {audio_file}: {e}")

# Ordenar por duración
filtered_files.sort(key=lambda x: x[1])

# Mostrar resultados
print(f"\nSe encontraron {len(filtered_files)} archivos de '{target_emotion}' con duración entre {min_duration} y {max_duration} segundos:")
for file_path, duration in filtered_files:
    print(f"{os.path.basename(file_path)}: {duration:.2f} segundos")

# Estadísticas
if filtered_files:
    durations = [d for _, d in filtered_files]
    avg_duration = sum(durations) / len(durations)
    min_dur = min(durations)
    max_dur = max(durations)

    print(f"\nEstadísticas de duración:")
    print(f"Promedio: {avg_duration:.2f} segundos")
    print(f"Mínima: {min_dur:.2f} segundos")
    print(f"Máxima: {max_dur:.2f} segundos")
```
